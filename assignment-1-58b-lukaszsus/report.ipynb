```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from settings import RESULTS_PATH
```

## Results

Every set of parameters was tested 20 times.

```python
results_file_path = os.path.join(RESULTS_PATH, "tables", "20200405_171633.csv")
df_results = pd.read_csv(results_file_path)
df_results
```

## Random classifier

```python
df_random = df_results[df_results["classifier"] == "random"]
fig = plt.figure(figsize=(20,15))
ax = sns.lineplot(data=df_random, x="uncovered_part", y="f1_macro", hue="uncovering", marker='o')
for line in ax.lines:
    line.set_linestyle("--")
ax.set(ylim=(0,100))
ax.set_title('Random classifier f1 score (macro average)')
```

## Iterative classification algorithm

```python
df_ica = df_results[df_results["classifier"] == "ica"]
fig = plt.figure(figsize=(20,15))
ax = sns.lineplot(data=df_ica, x="uncovered_part", y="f1_macro", hue="uncovering", marker='o')
for line in ax.lines:
    line.set_linestyle("--")
ax.set(ylim=(0,100))
ax.set_title('ICA f1 score (macro average)')
```

### Ica vs random

```python
df = df_results[df_results["uncovering"] == "betweenness"]
fig = plt.figure(figsize=(20,15))
ax = sns.lineplot(data=df, x="uncovered_part", y="f1_macro", hue="classifier", marker='o')
for line in ax.lines:
    line.set_linestyle("--")
ax.set(ylim=(0,100))
ax.set_title('ICA vs random f1 score (macro average)')
```
